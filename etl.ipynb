```python
import pandas as pd
import os
```

```python
directory = './resources/jersey_city'
csv_files = [f for f in os.listdir(directory) if f.endswith('.csv')]
csv_files
```

```python
df = pd.DataFrame()
header_added = False
```

```python
for file in csv_files:
    month_df = pd.read_csv(os.path.join(directory, file))

    if not header_added:
        df = month_df.copy()
        header_added = True
    else:
        df = pd.concat([df, month_df], ignore_index=True)
```

```python
df
```

```python
df.to_csv('./resources/JC_202110-202309.csv', index=False)
```
